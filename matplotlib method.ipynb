```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.bar(x,y,color='blue')
plt.title("bar Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
values = [2, 4, 6, 8, 10]
labels = ['A', 'B', 'C', 'D', 'E']  
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.axis('equal')  
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Add a label to the line for the legend
plt.plot(x, y, label="Line: y = 2x")

plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Now this will show the legend

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
x1 = [6, 7, 8, 9, 12]
y1 = [1, 2, 3, 4, 5]

# Plot first line
plt.plot(x, y, label="Line: y = 2x", color='blue')


plt.plot(x1, y1, label="Line: y = x1", color='green')

plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  

plt.show() 

```

```python
import matplotlib.pyplot as plt
import numpy as np  
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
x1 = [6, 7, 8, 9, 12]
y1 = [1, 2, 3, 4, 5]
x_array = np.array(x)
y_array = np.array(y)
z = x_array * y_array 
print("z =", z)
plt.plot(x, y, label="Line: y = 2x", color='blue')
plt.plot(x1, y1, label="Line: y = x1", color='green')
plt.plot(x, z, label="Line: z = x*y", color='red')
plt.title("Simple Line Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
